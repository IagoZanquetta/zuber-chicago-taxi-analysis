### Teste de Hipótese – Duração Média das Corridas em Sábados Chuvosos

**Objetivo:**
Verificar se a duração média das corridas do Loop para o Aeroporto Internacional O’Hare muda significativamente em sábados com condições meteorológicas ruins (indicadas por `"Bad"` na coluna `weather_conditions`).

---

### **Formulação das Hipóteses**

* **Hipótese Nula (H₀):**
  A média da duração das corridas é igual entre sábados chuvosos e os demais dias.
  μ₁ = μ₂

* **Hipótese Alternativa (H₁):**
  A média da duração das corridas é diferente em sábados chuvosos.
  μ₁ ≠ μ₂

---

### **Nível de Significância**

Foi adotado um nível de significância **α = 0.05 (5%)**, padrão em testes estatísticos.
Isso significa que aceitamos um erro de 5% ao rejeitar a hipótese nula.

---

### **Critério Estatístico**

Antes de aplicar o teste t de Student, é necessário verificar se as variâncias dos dois grupos (sábados chuvosos e demais dias) são iguais.
Para isso, será utilizado o **teste de Levene**, que avalia a igualdade das variâncias entre amostras.

* Se o **p-valor do teste de Levene > 0.05**, assumimos **variâncias iguais** e usamos `equal_var=True`.
* Se o **p-valor do teste de Levene ≤ 0.05**, as variâncias **são diferentes** e usamos `equal_var=False`.

Em seguida, será  aplicado o **teste t de Student para amostras independentes** (`scipy.stats.ttest_ind`), conforme o resultado do teste de Levene, para verificar se há diferença significativa entre as médias dos grupos.

```python
from scipy.stats import ttest_ind, levene

# Cria coluna para o dia da semana (0=segunda, ..., 6=domingo)
df_loop_ohare['weekday'] = df_loop_ohare['start_ts'].dt.weekday

# Verifica se é sábado
df_loop_ohare['is_saturday'] = df_loop_ohare['weekday'] == 5

# Considera "Bad" como condição chuvosa
df_loop_ohare['is_rain'] = df_loop_ohare['weather_conditions'].str.lower() == 'bad'

# Identifica sábados chuvosos
df_loop_ohare['is_rainy_saturday'] = df_loop_ohare['is_saturday'] & df_loop_ohare['is_rain']

# Separa os dois grupos
rainy_saturdays = df_loop_ohare[df_loop_ohare['is_rainy_saturday']]['duration_seconds']
other_days = df_loop_ohare[~df_loop_ohare['is_rainy_saturday']]['duration_seconds']

# Teste de Levene para verificar igualdade das variâncias ===
levene_stat, levene_p = levene(rainy_saturdays, other_days)
print(f"Estatística de Levene: {levene_stat:.3f}")
print(f"p-valor de Levene: {levene_p:.5f}")

# Define o parâmetro equal_var com base no resultado do teste de Levene
equal_var = True if levene_p > 0.05 else False
print(f"Variâncias iguais? {equal_var}")

# Aplica o teste t de Student conforme o resultado ===
t_stat, p_value = ttest_ind(rainy_saturdays, other_days, equal_var=equal_var)

print(f"\nEstatística t: {t_stat:.3f}")
print(f"p-valor: {p_value:.5f}")

# Interpretação
alpha = 0.05
if p_value < alpha:
    print("\nResultado: Rejeitamos a hipótese nula. Há diferença significativa na duração média das corridas em sábados chuvosos.")
else:
    print("\nResultado: Não rejeitamos a hipótese nula. Não há evidência de diferença significativa.")
```
